```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Employee_ID': [101, 102, 103, 104, 105, 102, 106, 107],
    'Name': ['Ali Khan', 'Sara Ahmed', ' John  Doe ', None, 'Ayesha Khan', 'Sara Ahmed', 'Imran Ali', '  sana malik '],
    'Age': [25, 30, 27, None, 22, 30, 999, 26],
    'Department': ['Sales', 'HR', 'IT', 'Finance', 'HR', 'HR', 'Sales', 'sales'],
    'Salary': ['50000', '$55000', '60000', '52000', 'missing', '$55000', '58000', '49000'],
    'Join_Date': ['2021-05-10', '05/12/2020', 'wrong_date', '2020-11-20', '2021-02-25', '05/12/2020', '2019-09-10', '2021-01-15'],
    'City': ['faisalabad', 'Lahore', 'Karachi', 'islamabad', 'Lahore', 'Lahore', 'Fsd', ' karachi']
}

df = pd.DataFrame(data)
df.to_csv('employees_dirty.csv', index=False)
print("✅ File 'employees_dirty.csv' created.")
```

# Check for missing values. Find how many missing values each column has.

```python
df.isnull() # that returns boolean value where True means null value and false means not null value
```

```python
df.isnull().sum()  # that tells the sum of null values in each coloumns
```

# Fill missing Age or Salary with mean/median.

```python
# Remove $ sign and 'missing' text
df['Salary'] = df['Salary'].replace('[$,]', '', regex=True)
df['Salary'] = df['Salary'].replace('missing', None)

# Convert to numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
```

```python
df['Age'].mean()
```

```python
df.loc[df['Age'].isnull(), 'Age'] = df['Age'].mean()
df.loc[df['Salary'].isnull(), 'Salary'] = df['Salary'].mean()
```

```python
df.drop_duplicates(inplace=True,keep='first')
```

```python
df['Join_Date']=pd.to_datetime(df['Join_Date'],errors='coerce')
df
```

# Correct Wrong Ages (like 999)

```python
df.loc[df['Age']>100, 'Age']=df['Age'].median()

df
```

```python
from datetime import datetime
# Fill invalid Join_Date (NaT) with today’s date
df['Join_Date'].fillna(pd.Timestamp(datetime.today().date()), inplace=True)

# Remove spaces from Name
df['Name'] = df['Name'].str.strip()

# Capitalize Department & City
df['Department'] = df['Department'].str.title()
df['City'] = df['City'].str.title()

# Add Experience column
today = pd.Timestamp(datetime.today().date())
df['Experience (years)'] = (today - df['Join_Date']).dt.days / 365
df['Experience (years)'] = df['Experience (years)'].round(1)

# Verify results
print(df.head())
print(df.info())

# Average Salary by Department
avg_salary = df.groupby('Department')['Salary'].mean().reset_index()
print("\nAverage Salary by Department:")
print(avg_salary)
```
