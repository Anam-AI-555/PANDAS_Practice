# Replace all missing values in numeric columns with the mean of that column

```python
# Fill missing numeric values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)
df

# df.fillna(df.mean(numeric_only=True),inplace=True)
```

# Convert Sale_Date to datetime

```python
df['Sale_Date ']=pd.to_datetime(df['Sale_Date'])


print(df['Sale_Date'].head(10))
print(df['Sale_Date'].dtype)
```

# Check for missing values in numeric columns (Sales_Amount, Unit_Cost, Unit_Price, Discount)

```python
df[['Sales_Amount','Unit_Cost','Unit_Price','Discount']].isnull().sum()
```

# Fill missing discounts with 0

```python
df['Discount'].fillna(0,inplace=True)
df.head()
```

# Check unique values in categorical columns (Region, Product_Category, Payment_Method, etc.)

```python
for col in ['Region','Product_Category','Sales_Rep','Customer_Type','Payment_Method','Region_and_Sales_Rep']:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```

# Find top 5 sales reps by total Sales_Amount(

```python
df.groupby('Sales_Rep')['Sales_Amount'].sum().sort_values(ascending=False)
df.head(10)
```

# Add a new column Profit = Sales_Amount - (Unit_Cost × Quantity_Sold)

```python
df['profit']=df['Sales_Amount']-(df['Unit_Cost']*df['Quantity_Sold'])
df['profit'].head()
```

# Add a column for net sales after discount

```python
df['net_sales']=df['Sales_Amount']-df['Discount']
df.head()
```

# Top product categories by sales

```python
df.groupby('Product_Category')['Sales_Amount'].sum().sort_values(ascending=False)
```
